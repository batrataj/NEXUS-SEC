```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/batrataj/NEXUS-SEC/main/Wednesday-workingHours.pcap_ISCX.xlsx"

df = pd.read_excel(url)

df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df['Label'].value_counts()
```

```python
df.to_csv("clean_network_data.xlsx", index=False)
```
